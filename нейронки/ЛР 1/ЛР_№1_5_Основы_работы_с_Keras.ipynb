Таким образом, когда мы решаем пошевелить руками, сигналы от мозга передаются мышцам наших рук, а не мышцам ног.

# Обучение нейронной сети

После понимания различных признаков искусственной нейронной сети остается один вопрос: как сеть «обучается»?

Возьмем, к примеру, молодого баскетболиста, который учится бросать трехочковый бросок. Если он бросает и промахивается из-за того, что полет мяча слишком короткий, баскетболист исправляется и увеличивает силу следующего броска. Если следующий бросок сейчас будет слишком далеко справа от корзины, игрок снова изменит свой бросок, чтобы кидать больше в центр. Игрок продолжает делать это до тех пор, пока не будет произведен точный бросок. Затем игрок запоминает точную силу и направление броска при выполнении трехочковых в будущем.

Это похоже на обучение нейронных сетей. Сначала данные передаются по сети, и прогнозируемый результат дает n. Это известно как прямое распространение. Затем прогнозируемые выходные данные сравниваются с фактическими выходными данными, и различия между прогнозируемыми и фактическими данными передаются обратно через модель. Во время обратного прохода в модель будут внесены корректировки, так что различия между выходными данными прогноза и фактическими выходными данными будут уменьшены. Это известно как обратное распространение. После внесения корректировок данные будут снова передаваться из входного слоя, и будет сделан другой прогнозируемый вывод. Новый прогнозируемый результат снова будет сравниваться с фактическим, а различия будут переданы обратно через модель. В модель будут внесены дополнительные корректировки.

Процесс прямого и обратного распространения будет повторяться до тех пор, пока различия между прогнозируемым выходом и фактическим выходом не будут сведены к минимуму. Теперь модель обучена и может использоваться для прогнозирования других подобных наборов данных.

Попробуйте использовать приведенный ниже код, чтобы создать искусственную нейронную сеть.

```python
#  Инициализируем нейронную сеть 
model = Sequential()

# Добавим первый скрытый слой с 6 узлами. Input_dim относится к количеству столбцов / количеству признаков в x_values или входном слое.
# Активация относится к тому, как активируются узлы / нейроны. Мы будем использовать relu. Другие распространенные активации - сигмовидная и гиперболический тангенс (tanh)
model.add(Dense(6,input_dim=4,activation='relu'))

# Добавим скрытый слой с 6 узлами. 
model.add(Dense(6,activation='relu'))

# Добавляем выходной слой с 3 узлами. Используемая активация должна быть softmax. Softmax используется, когда вы имеете дело с категориальными выходами или целями. 
model.add(Dense(3,activation='softmax'))

# Скомпилируйте модель вместе. Оптимизатор относится к методу корректировки в модели. Loss относится к разнице между прогнозируемыми и фактическими значениями
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

Опционально: [Сравнения между функциями активации](https://neurohive.io/ru/osnovy-data-science/activation-functions/). Вы узнали о ReLu ранее на этапе Приобретения опыта. Мы можем использовать функцию to_categorical из Keras. Попробуйте использовать приведенный ниже код, чтобы пометить кодировку, а затем унитарно закодируйте y_values.


```python
from keras.utils import to_categorical

# Выведем количество точек для кажлого класса
print(df['class'].value_counts())

# Словарь для кодирования классов. 
# Мы используем значения, начиная с 0, поэтому при унитарном кодировании будут созданы только 3 столбца вместо 4.
label_encode = {"class": {"Iris-setosa":0, "Iris-versicolor":1, "Iris-virginica":2}}

# Используем .replace чтобы преобразовать разные классы в числа
df.replace(label_encode,inplace=True)

# Выведем количество точек данных в каждом классе, чтобы проверить, изменились ли классы на числа
print(df['class'].value_counts())

# Извлечем класс как y_values
y_values = df['class']

# Проведем кодирование y_values
y_values = to_categorical(y_values)

print(y_values)
```

Посмотрите, что указано в y_values выше. Мы закодировали названия цветов в числа.

```python
# Инициализируем нейронную сеть как модель
model4 = Sequential()

# Добавим первый скрытый слой с 6 узлами. Input_dim относится к количеству столбцов / количеству функций в x_values или input_layer.
# Активация относится к тому, как активируются узлы / нейроны. Мы будем использовать relu.
model4.add(Dense(6,input_dim=4,activation='relu'))

# Добавим скрытый слой с 6 узлами.
model4.add(Dense(6,activation='relu'))

# Добавляем выходной слой с 3 узлами. Используемая активация должна быть softmax. Softmax используется, когда вы имеете дело с прогнозными результатами или целями. 
model4.add(Dense(3,activation='softmax'))

# Скомпилируйте модель вместе.  
model4.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

# Обучить модель с помощью x_values и y_values. Эпохи относятся к тому, сколько раз полный набор данных будет использоваться для обучения модели.
# Shuffle = True сообщает модели, что необходимо рандомизировать порядок набора данных после каждой эпохи. Это позволит модели учиться.
model4.fit(x_values,y_values,epochs=200,shuffle=True)
```

Как видно, показатель точности теперь превышает 90%. Вот это да! Кажется удивительным, что простое увеличение количества эпох позволяет повысить точность. Как Вы думаете, хорошо ли, что модель настолько точна на данных, на которых она обучалась? Только представьте, сможет ли футболист преуспеть в матче, если он / она будет очень усердно тренироваться, чтобы забить гол с определенного места на поле? Или пекарь сможет испечь вкусный торт по запросу клиента, если он / она научится печь только торт определенного вкуса, формы и размера?

Футболист не сможет хорошо выступить в матче, так как он / она не сможет точно бить из других частей поля. Пекарь не сможет испечь вкусный торт, так как знает только один конкретный вкус.

Идея этого вопроса заключается в концепции переобучения. Если модель переоценивается, она не сможет обобщить, чтобы правильно спрогнозировать данные, которых она раньше не видела.

Это концепция переобучения, которая также применима ко всем методам машинного обучения. Как только методика будет слишком точно соответствовать набору данных, обученная методика не сможет обобщить другие данные, которые она не видела раньше. Таким образом, мы обычно обучаем модель только части набора данных, который у нас есть, а оставшуюся часть оставляем в качестве набора для тестирования или проверки, чтобы увидеть, не переобучена ли модель. При обучении на обучающей выборке точность модели на тестовой должна возрасти. Однако в точке переобучения точность тестового набора начнет снижаться. Если Вы видите, что точность тестирования начала увеличиваться после определенной эпохи, вам не следует обучать модель дальше.

Во-первых, нам нужно решить, сколько данных мы хотим приберечь на тестирование. Обычно мы забираем от 20% до 30% набора данных. В этом примере мы оставим 25 процентов набора данных в качестве набора для тестирования / проверки. Мы можем использовать функцию train_test_split из библиотеки sklearn. Попробуйте приведенный ниже код.

```python
from sklearn.model_selection import train_test_split

# Извлечем исходные значения x_values из df.
# Мы должны повторно извлечь значения x_values, поскольку стандартизация должна основываться только на данных, с которыми будет обучаться модель.
# Таким образом, мы должны сначала разделить данные перед стандартизацией.

x_values = df[['sepal_length','sepal_width','petal_length','petal_width']]

# Test_size = 0,25 указывает, что 25% точек данных будут в x_test и y_test, тогда как 75% будут в x_train и y_train.
# random_state = 10 используется для обеспечения того, чтобы разделение было одинаковым каждый раз, когда вы запускаете приведенный ниже код.
# Это потому, что разбиение выполняется каждый раз случайным образом. Одно и то же random_state - единственный способ обеспечить каждый раз одно и то же разбиение.
x_train, x_test, y_train, y_test = train_test_split(x_values,y_values,test_size=0.25,random_state=10)

# Проверяем количество строк в x_train, x_test, y_train и y_test
print("Number of rows in x_train:", x_train.shape[0])
print("Number of rows in x_test:", x_test.shape[0])
print("Number of rows in y_train:", y_train.shape[0])
print("Number of rows in y_test:", y_test.shape[0])

# Теперь мы можем стандартизировать значения x.
# Инициализируем StandardScaler

standardise = StandardScaler()

# Стандартизируем значения x_train с помощью .fit_transform
x_train = standardise.fit_transform(x_train)

# Стандартизируем значения x_test с помощью .transform. 
# Нет необходимости подбирать данные, поскольку стандартизация должна быть такой же, как у x_train.
x_test = standardise.transform(x_test)
```
