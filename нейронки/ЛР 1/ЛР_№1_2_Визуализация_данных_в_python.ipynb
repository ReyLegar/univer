## Точечный график (диаграмма рассеяния)

Диаграмма рассеяния используем точки графика, чтобы представить значения двух переменных. Это позволяет увидеть наличие каких-либо зависимостей сежду этими двумя переменными. 

Давайте посмотрим, как меняется средняя температура с годами! 

```python
x = dataframe.date
y = dataframe.mean_temperature
```

```python
plt.scatter(x,y)
plt.show()
```

Вы видите, что ось Х нарисована толстой линией, а отметки не видны? Это лишает нас возможности анализировать данные. 

Давайте видоизменим нашу диаграмму,сделав ее более читаемой с точки зрения оси Х!

### Используем не все значения Х для визуализации
Одна из причин, почему у нас получилась толстая линия под графиком - это большое количество добписей к оси Х  (2 года ежедневных данных). 

Первое, что мы сделаем, это уменьшим количество меток по оси Х. Сделаем это с помощью метода np.arrange:

```python
plt.scatter(x,y)
plt.xticks(np.arange(0, 731, 30))
plt.show()
```

Видите, теперь значения по оси Х стали различимее, но все еще перекрывают друг друга.

### Рисование нескольних графиков на одной диаграмме

```python
x = dataframe.date
y_1 = dataframe.max_temperature
y_2 = dataframe.min_temperature

plt.plot(x,y_1, label = "Max temp") 
plt.plot(x,y_2, label = "Min temp") 

plt.xticks(np.arange(0, 731, 60))
plt.xticks (rotation=30)

plt.legend()
plt.show()
```

### Задание 16: Нарисуйте на диаграмме не менее 3 графиков!

Мы рассмотрим диаграмму размаха, используя образец руководства, полученный с  [веб-сайта](https://matplotlib.org/gallery/pyplots/boxplot_demo_pyplot.html#sphx-glr-gallery-pyplots-boxplot-demo-pyplot-py) matplotlib.

## Подграфики

Часто Вам нужно построить несколько графиков рядом. Для этого вы можете использовать подграфики!

Вот как они делаются!

```python
x = dataframe.date
y = dataframe.mean_temperature

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True) #Создаем 2 подграфика с общей осью y 
fig.suptitle('Общая ось Y')
ax1.plot(x, y)
ax2.scatter(x, y)
plt.show()
```
